```python
arc = 'data/archive'
# Pulls subdirectories out of the archive directory
subs = sorted(next(os.walk(arc))[1])
```

```python
# Remove the directories to be copied into and the one that looks like a photo name
desired_subs = subs[:-2]
```

```python
# change values to numbers
for i in range(0, len(desired_subs)):
    desired_subs[i] = int(desired_subs[i]) 
```

```python
desired_subs = sorted(desired_subs)
```

```python
# first_few = minus_others[:3]
```

```python
# copy photos into IDC and Non IDC directories
idc = 0
non = 0
for num in desired_subs:
    path = 'data/archive/{}'.format(num)
    path_idc = path + '/1'
    path_non = path + '/0'
#     paths.append(path_non)

    for img in os.listdir(path_idc):
        image = os.path.join(path_idc, img)
#         shutil.copy2(image, 'data/archive/idc/') 
        idc += 1
        
    for img in os.listdir(path_non):
        image = os.path.join(path_non, img)
#         shutil.copy2(image, 'data/archive/non_idc')
        non += 1
```
